# BOXPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("penguins")
df.head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="species", y="flipper_length_mm", hue="species", data=df, palette="Set3", legend=False)
plt.title("Penguen Türlerine Göre Flipper Uzunluklarının Dağılımı",fontsize=14)
plt.xlabel("Penguen Türü", fontsize=12)
plt.ylabel("Flipper Uzunluğu (mm)", fontsize=12)
plt.show()
```

# Line Chart 

```python
import pandas as pd
df = pd.read_csv('C:\\Users\\yusuf\\Desktop\\StudentsPerformance.csv')
df.head()
```

```python
df_grouped = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_grouped, x='race/ethnicity', y='math score', marker='o', label='Matematik', color='blue')
sns.lineplot(data=df_grouped, x='race/ethnicity', y='reading score', marker='o', label='Okuma', color='green')
sns.lineplot(data=df_grouped, x='race/ethnicity', y='writing score', marker='o', label='Yazma', color='orange')

plt.title('Gruplarının Ders Başarıları Karşılaştırması', fontsize=14)
plt.xlabel('Grup (race/ethnicity)', fontsize=12)
plt.ylabel('Ortalama Puan', fontsize=12)

plt.legend(title='Dersler')

plt.show()
```

# Bar Chart

```python
import pandas as pd
df = pd.read_csv('C:/Users/yusuf/Desktop/HRDataset_v14.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df_grouped = df.groupby('Department').size().reset_index(name='Employee Count')

plt.figure(figsize=(12, 6))
sns.barplot(data=df_grouped, x='Department', y='Employee Count', hue='Department', palette='Set2', legend=False)

plt.title('Departman Bazında Çalışan Sayısı Dağılımı', fontsize=14)
plt.xlabel('Departman', fontsize=12)
plt.ylabel('Çalışan Sayısı', fontsize=12)

plt.show()
```

# Violin graph

```python
import pandas as pd
df = pd.read_csv('C:/Users/yusuf/Desktop/melb_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9,6)) 
sns.violinplot(x='Regionname', y='Price', data=df, inner='quart', palette='Set3', hue='Regionname', legend=False)

plt.xticks(rotation=45, fontsize=12) 

plt.title('Bölgeler Bazında Fiyat Dağılımı', fontsize=14)
plt.xlabel('Bölge', fontsize=12)
plt.ylabel('Fiyat (AUD)', fontsize=12)

plt.show()
```
